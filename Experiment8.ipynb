```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits

dat=load_digits()
df=pd.DataFrame(dat.data)

X=df
y=pd.Series(dat.target)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler=StandardScaler()
X=scaler.fit_transform(X)

xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC

model_linear=SVC(kernel='linear')
model_linear.fit(xtrain,ytrain)
linear_ypred=model_linear.predict(xtest)


model_rbf=SVC(kernel='rbf')
model_rbf.fit(xtrain,ytrain)
rbf_ypred=model_rbf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_li=accuracy_score(ytest,linear_ypred)
accuracy_rbf=accuracy_score(ytest,rbf_ypred)

print("accuracy before tuning")
print("accuracy linear: ",accuracy_li)
print("accuracy rbf: ",accuracy_rbf)
```

```python
from sklearn.model_selection import GridSearchCV

grid1={
    "C":[0.1,1,10,100],
    "kernel":['linear']
}

grid2={
    "C":[1,10,100],
    "gamma":['scale',0.1,0.01],
    "kernel":['rbf']
}

model_li=GridSearchCV(SVC(),grid1,cv=5)
model_li.fit(xtrain,ytrain)
results1=model_li.best_params_

model_rbf=GridSearchCV(SVC(),grid2,cv=5)
model_rbf.fit(xtrain,ytrain)
results2=model_rbf.best_params_

```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

def accuracy_tuned(results):
  model_tuned = SVC(**results)
  model_tuned.fit(xtrain, ytrain)
  tuned_ypred = model_tuned.predict(xtest)
  acc_tuned = accuracy_score(ytest,tuned_ypred)
  print(f"Tuned accuracy {results["kernel"]}: ", acc_tuned)


print("Accuracy after tuning")
accuracy_tuned(results1)
accuracy_tuned(results2)
```

```python
print("Best parameters for linear",results1)
print("Best parameters for rbf",results2)
```
